```python
num_var = ['is_canceled', 'lead_time', 'arrival_date_year',
       'arrival_date_week_number', 'arrival_date_day_of_month',
       'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children',
       'babies', 'is_repeated_guest', 'previous_cancellations',
       'previous_bookings_not_canceled', 'booking_changes',
       'days_in_waiting_list', 'adr', 'required_car_parking_spaces',
       'total_of_special_requests']

plt.figure(figsize=(25,25))
for i,var in enumerate(num_var):
    plt.subplot(9,4,i+1)
    sns.distplot(data[var],bins=20)
    sns.distplot(data3_drop_rows[var],bins=20)
```

```python
data3_drop_rows['hotel'].value_counts()/data3_drop_rows.shape[0] * 100
```

```python
pd.concat([data['hotel'].value_counts()/data.shape[0] * 100,
          data3_drop_rows['hotel'].value_counts()/data3_drop_rows.shape[0] * 100],axis = 1,
         keys = ["hotel_org","hotel_clean"])
```

```python
def cat_var_dist(var):
    return pd.concat([data[var].value_counts()/data.shape[0] * 100,
              data3_drop_rows[var].value_counts()/data3_drop_rows.shape[0] * 100],axis = 1,
                  keys = [var+"_org",var+"_clean"])
```

```python
cat_var_dist("hotel")
```

```python
cat_var = ['hotel', 'arrival_date_month', 'meal', 'country', 'market_segment',
       'distribution_channel', 'reserved_room_type', 'assigned_room_type',
       'deposit_type', 'customer_type', 'reservation_status',
       'reservation_status_date', 'name', 'email', 'phone-number',
       'credit_card']

for var in cat_var:
    print(cat_var_dist(var))
```
