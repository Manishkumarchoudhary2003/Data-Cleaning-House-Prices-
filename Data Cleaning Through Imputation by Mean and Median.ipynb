```python
null_var = df.isnull().sum()/df.shape[0] * 100
```

```python
null_var
```

```python
drop_columns = null_var[null_var > 20].keys()
print(drop_columns)
```

```python
df2_drop_clm = df.drop(columns= drop_columns)
```

```python
df2_drop_clm.shape
```

```python
df3_num = df2_drop_clm.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
plt.figure(figsize=(25,25))
sns.heatmap(df3_num.isnull())
```

```python
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
missing_num_var = [var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
```

```python
missing_num_var
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws = {"linewidth":5,"color":"#DC143C"})
```

```python
df4_num_mean = df3_num.fillna(df3_num.mean())
```

```python
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws = {"linewidth":5,"color":"#DC143C"},label="org")
    sns.distplot(df4_num_mean[var],bins=20,kde_kws = {"linewidth":5,"color":"#DC143C"},label="Mean")
    plt.legend()
```

```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False,kde_kws = {"linewidth":5,"color":"#DC143C"},label="org")
    sns.distplot(df4_num_mean[var],bins=20,hist=False,kde_kws = {"linewidth":5,"color":"#DC143C"},label="Mean")
    sns.distplot(df5_num_median[var],bins=20,hist=False,kde_kws = {"linewidth":5,"color":"#DC143C"},label="Median")
    plt.legend()
```

```python
df_concat = pd.concat([df[missing_num_var],df4_num_mean[missing_num_var],df5_num_median[missing_num_var]])
```
